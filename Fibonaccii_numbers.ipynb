```python
import numpy as np  # for numericall operations
import matplotlib.pyplot as plt # for creating visualizations

# Function to generate Fibonacci numbers
def fibonacci(n):
    fib_sequence = [0, 1]  # initialize with the frst two fibonacci numbers
    # calculating each new Fibonacci number by adding the previous two
    for i in range (2, n):  # loop from index 2 to n-1
        fib_sequence.append(fib_sequence[i-1] + fib_sequence[i-2])
    return fib_sequence  # return the complete sequence

# Set the number of Fibonacci numbers to generate
n = 100  # We want to first 100 numbers

# Call the function to generate the sequence
fib_numbers = fibonacci(n) # Store the result in fib_numbers

#Create a new figure with specified size 
plt.figure(figsize=(12, 6))  # Width=12 inches, height=6 inches

# Plot the Fibonacci sequence as a blue line 
plt.plot(range(n), fib_numbers, 'b-', linewidth=1.5)  # 'b-' means blue line, linewidth controls thickness

# add red dots at each Fibonacci number position
plt.scatter(range(n), fib_numbers, c='red', s=30, alpha=0.7)  # c=color, s=size, alpha=transparency

# Add a title yo yhe plot
plt.title('First 100 Fibonacci Numbers', fontsize=16)

#Label the x-axis
plt.xlabel('Index', fontsize=12)  #Describes what the x-axis represent

# Label the y-axis
plt.ylabel('Value', fontsize=12)   #Describes what the y-axis represent

# Add a grid to make the plot easier to read
plt.grid(True, alpha=0.3)

# Set y-axis to logarithmic scale to better visualize exponential growth
plt.yscale('log')  # Without this later values would make earlier ones invisible

# Adjust the layout to ensure everything fits well
plt.tight_layout()  #Prevents labels from being cut off

# Display the plot 
plt.show()  #This command renders the visualization

# Print the first 10 numbers for reference
print("First 10 Fibonacci Numbers:", fib_numbers[:10])  

# Print the last 5 numbers for reference
print("Last 5 Fibonacci Numbers:", fib_numbers[-5:])
```
